```python
import pandas as pd
```

```python
from pycaret.regression import *
```

```python
df = pd.read_csv('data/nba_2023_dataset.csv')

s = setup(df,target='WL')
```

```python
best = compare_models()
```

```python
predictions = predict_model(best)
```

```python
# Testing model with new dataset
dfa = pd.read_csv('data/nba_game_data_020124.csv') 
predictions_test = predict_model(best,data=dfa)
predicted = predictions_test[predictions_test['TEAM_NAME']=='Golden State Warriors']
predicted_filtered= predicted[['TEAM_NAME','MATCHUP','prediction_label']]

predicted_filtered = predicted_filtered.drop_duplicates(subset=['MATCHUP'])
print(predicted_filtered.to_string())
```
